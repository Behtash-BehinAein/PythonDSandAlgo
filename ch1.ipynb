# Generators

#### Example: factors of n

```python
def factors(n):
    for k in range(1,n+1):
        if n%k==0:
            yield k
```

```python
for f in factors(15):
    print(f)
```

#### Example: factors of n (more efficient)

```python
def factors(n):
    k=1
    while k**2 < n:
        if n%k == 0:
            yield k, n//k
        k += 1
    if k**2 ==n:
        yield k
        
```

```python
for a in factors(15):
    print(a)
```

#### Example: infinite fib series

```python
def fib():
    a, b = 0 , 1
    while True:
        yield a
        next_val = a + b
        a = b
        b = next_val
```

```python
stop = 10
cc = 0
for i in fib():
    print(i)
    cc+=1
    if cc>stop: break
```

```python
def minmax(seq):
    biggest  = seq[0]
    smallest = seq[0]
    for num in seq:
        if num > biggest:
            biggest = num
        if num < smallest:
            smallest = num
    return smallest , biggest
```

```python
import random as rnd
seq = [chr(i) for i in range(97,123)]

def shuff(seq):
    indices = list( range(len(seq)) )
    while len(indices)>=3:
        l , r  = indices.pop(0)  , indices.pop(rnd.randint(1 , len(indices)-1)) 
        seq[l] , seq[r] = seq[r] , seq[l] 
    return seq
    
```

```python
def factors(n):
    k=1
    lst = []
    while k**2 < n:
        if n%k == 0:
            yield k
            lst.append(n//k)
        k += 1
    if k**2 ==n:
        yield k
    for i in lst[-1::-1]:
        yield i 
        
```
